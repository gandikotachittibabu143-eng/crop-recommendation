```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import numpy as np

# Load the trained model
model = joblib.load('crop_recommendation_model.pkl')

st.title('Crop Recommendation System')
st.write('Enter the environmental parameters to get a crop recommendation.')

# Input fields for the features
N = st.slider('Nitrogen (N) (kg/ha)', 0, 140, 50)
P = st.slider('Phosphorus (P) (kg/ha)', 0, 145, 50)
K = st.slider('Potassium (K) (kg/ha)', 0, 205, 50)
temperature = st.slider('Temperature (°C)', 0.0, 45.0, 25.0)
humidity = st.slider('Humidity (%)', 0.0, 100.0, 70.0)
ph = st.slider('pH Value', 0.0, 14.0, 6.0)
rainfall = st.slider('Rainfall (mm)', 0.0, 300.0, 100.0)
soil_organic_carbon = st.slider('Soil Organic Carbon (%)', 0.0, 10.0, 1.5) # New input for Soil Organic Carbon

# Create a DataFrame from inputs
# Note: soil_organic_carbon is included in the input_data for future use,
# but the current model (RandomForestClassifier) was trained without this feature.
# It will be ignored by the model for prediction purposes if not part of the training features.
input_data = pd.DataFrame([[N, P, K, temperature, humidity, ph, rainfall]],
                            columns=['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall'])

# --- Feature Importance Visualization ---
st.subheader('Feature Importance from the Model')
feature_names = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
feature_importances = model.feature_importances_

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

fig_importance, ax_importance = plt.subplots(figsize=(10, 6))
ax_importance.bar(importance_df['Feature'], importance_df['Importance'], color='skyblue')
ax_importance.set_xlabel('Features')
ax_importance.set_ylabel('Importance Score')
ax_importance.set_title('Feature Importance for Crop Recommendation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
st.pyplot(fig_importance)

# --- Crop Recommendation ---
if st.button('Get Recommendation'):
    # Get prediction probabilities for all classes
    probabilities = model.predict_proba(input_data)

    # Get class names (crop labels)
    class_names = model.classes_

    # Create a DataFrame of probabilities
    prob_df = pd.DataFrame({'Crop': class_names, 'Probability': probabilities[0]})
    prob_df = prob_df.sort_values(by='Probability', ascending=False)

    st.subheader('Top Crop Recommendations')

    # Display top N recommendations
    N_top_crops = 5 # You can change N here
    for index, row in prob_df.head(N_top_crops).iterrows():
        st.write(f"**{row['Crop']}**: {row['Probability'] * 100:.2f}%")

    # Display the single most recommended crop (highest probability)
    st.success(f'The most recommended crop is: **{prob_df.iloc[0]["Crop"]}** with a probability of **{prob_df.iloc[0]["Probability"] * 100:.2f}%**')
```

## Refine UI Layout and Styling

### Subtask:
Improve the visual organization and aesthetics of the Streamlit application. This could include using `st.sidebar` for inputs, `st.columns` for better arrangement of outputs, and adding more descriptive text for user guidance.

#### Instructions
1. Import the `streamlit` library if not already imported.
2. Modify the `app.py` file to move all input sliders ('N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall', and 'soil_organic_carbon') into `st.sidebar`.
3. Add a descriptive title and introductory text within the main content area of the Streamlit app.

**Reasoning**:
To implement the UI improvements, I will modify the `app.py` file. This involves moving input sliders to the sidebar, adding descriptive text, and using `st.columns` for better layout of the output sections.



```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import numpy as np

# Load the trained model
model = joblib.load('crop_recommendation_model.pkl')

st.set_page_config(layout="wide") # Set page layout to wide for better use of space

st.title('🌾 Smart Crop Recommendation System 🌾')
st.write('Welcome to the Smart Crop Recommendation System! This application helps farmers and agricultural enthusiasts identify the most suitable crops for their land based on various environmental parameters. Simply adjust the sliders in the sidebar to input your soil and climate conditions, and the system will provide tailored recommendations.')
st.markdown('---')

# --- Sidebar for Input Fields ---
st.sidebar.header('Environmental Parameters Input')

# Input fields for the features moved to sidebar
N = st.sidebar.slider('Nitrogen (N) (kg/ha)', 0, 140, 50, help="Amount of Nitrogen in the soil")
P = st.sidebar.slider('Phosphorus (P) (kg/ha)', 0, 145, 50, help="Amount of Phosphorus in the soil")
K = st.sidebar.slider('Potassium (K) (kg/ha)', 0, 205, 50, help="Amount of Potassium in the soil")
temperature = st.sidebar.slider('Temperature (°C)', 0.0, 45.0, 25.0, help="Average ambient temperature")
humidity = st.sidebar.slider('Humidity (%)', 0.0, 100.0, 70.0, help="Relative humidity")
ph = st.sidebar.slider('pH Value', 0.0, 14.0, 6.0, help="Soil pH level (acidity/alkalinity)")
rainfall = st.sidebar.slider('Rainfall (mm)', 0.0, 300.0, 100.0, help="Annual rainfall in millimeters")
soil_organic_carbon = st.sidebar.slider('Soil Organic Carbon (%)', 0.0, 10.0, 1.5, help="Percentage of organic carbon in the soil (for future model enhancements)") # New input for Soil Organic Carbon

st.sidebar.markdown('---')
st.sidebar.info('Adjust the sliders above to see crop recommendations tailored to your specific environmental conditions.')

# Create a DataFrame from inputs
input_data = pd.DataFrame([[N, P, K, temperature, humidity, ph, rainfall]],
                            columns=['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall'])

# --- Crop Recommendation and Visualizations in Main Area ---
st.subheader('Analyze & Get Recommendations')

if st.button('Get Recommendation', key='recommend_button'):
    # Get prediction probabilities for all classes
    probabilities = model.predict_proba(input_data)

    # Get class names (crop labels)
    class_names = model.classes_

    # Create a DataFrame of probabilities
    prob_df = pd.DataFrame({'Crop': class_names, 'Probability': probabilities[0]})
    prob_df = prob_df.sort_values(by='Probability', ascending=False)

    # Use columns for layout
    col1, col2 = st.columns(2)

    with col1:
        st.subheader('Top Crop Recommendations')
        N_top_crops = 5 # You can change N here
        for index, row in prob_df.head(N_top_crops).iterrows():
            st.write(f"**{row['Crop']}**: {row['Probability'] * 100:.2f}%")

        # Display the single most recommended crop (highest probability)
        st.success(f'The most recommended crop is: **{prob_df.iloc[0]["Crop"]}** with a probability of **{prob_df.iloc[0]["Probability"] * 100:.2f}%**')

    with col2:
        st.subheader('Feature Importance from the Model')
        feature_names = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
        feature_importances = model.feature_importances_

        importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
        importance_df = importance_df.sort_values(by='Importance', ascending=False)

        fig_importance, ax_importance = plt.subplots(figsize=(10, 6))
        ax_importance.bar(importance_df['Feature'], importance_df['Importance'], color='lightgreen')
        ax_importance.set_xlabel('Features')
        ax_importance.set_ylabel('Importance Score')
        ax_importance.set_title('Feature Importance for Crop Recommendation')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        st.pyplot(fig_importance)

st.markdown('---')
st.markdown('Developed with ❤️ for agricultural insights.')
```
